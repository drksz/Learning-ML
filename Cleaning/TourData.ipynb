```python
concerts = concerts.drop(labels=["Peak", "All Time Peak"], axis=1)
```

```python
for col in concerts.columns:
    print(repr(col))
```

```python
concerts.columns = (
    concerts.columns
    .str.replace("\xa0", " ", regex=False)
    .str.replace(" ", "_", regex=False)
)
```

```python
concerts.columns
```

```python
concerts.head(20)
```

```python
concerts['Actual_gross'] = concerts['Actual_gross'].str.replace(r"\D+","", regex=True)
concerts['Adjusted_gross_(in_2022_dollars)'] = concerts['Adjusted_gross_(in_2022_dollars)'].str.replace(r"\D+","", regex=True)
concerts['Actual_gross'] = concerts['Actual_gross'].astype(int)
concerts['Adjusted_gross_(in_2022_dollars)'] = concerts['Adjusted_gross_(in_2022_dollars)'].astype(int)
```

```python
concerts.info()
```

```python
concerts['Tour_title'] = (
    concerts['Tour_title']
    .str.replace(r"\[.*\]", "", regex=True)
    .str.replace(r"[^\w\s:]", "", regex=True)
    .str.strip()
)

concerts['Ref.'] = (
    concerts['Ref.']
    .str.replace("d", "5")
    .str.replace(r"[^0-9]","", regex=True)
    .str.replace("1516", "15")
)
```

```python
concerts['Average_gross'] = concerts['Average_gross'].str.replace(r"\D","", regex=True)
concerts['Average_gross'] = concerts['Average_gross'].astype(int)
```
